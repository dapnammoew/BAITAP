BÀI 3

```python
a1, b1 = np.array([1, 5, 9]), np.array([2, 5, 8])
print("1D Greater:", np.greater(a1, b1))
print("1D Greater Equal:", np.greater_equal(a1, b1))
print("1D Less:", np.less(a1, b1))
print("1D Less Equal:", np.less_equal(a1, b1))

# So sánh 2 chiều
a2, b2 = np.array([[1, 2], [3, 4]]), np.array([[1, 3], [2, 4]])
print("\n2D Greater:\n", np.greater(a2, b2))
print("2D Greater Equal:\n", np.greater_equal(a2, b2))
print("2D Less:\n", np.less(a2, b2))
print("2D Less Equal:\n", np.less_equal(a2, b2))
```

BÀI 19

```python
# Bài 19: Tìm những phần tử trong Series x mà không có trong Series y (hàm isin)
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

# Dùng toán tử ~ để lấy phần bù của hàm isin
diff = sr1[~sr1.isin(sr2)]
print("Items of sr1 not present in sr2:\n", diff)
```

BÀI 20

```python
# Bài 20: Phép toán numpy.union1d, numpy.intersect1d (hội, giao)
# trên 2 Series ngẫu nhiên có 20 phần tử
sx = pd.Series(np.random.randint(1, 50, size=20))
sy = pd.Series(np.random.randint(1, 50, size=20))

print("Hội (Union):", np.union1d(sx, sy))
print("Giao (Intersection):", np.intersect1d(sx, sy))
```

BÀI 21

```python
# Bài 21: Thêm series y vào series x theo chiều dọc, chiều ngang
# (Hàm .append() của Series đã bị gỡ bỏ ở phiên bản mới, ta dùng pd.concat)
series_x = pd.Series([0, 1, 2, 7, 8, 9])
series_y = pd.Series(['p', 'q', 'r', 'w', 'x', 'y'])

print("Chiều dọc (Vertical):\n", pd.concat([series_x, series_y], axis=0, ignore_index=True))
print("\nChiều ngang (Horizontal):\n", pd.concat([series_x, series_y], axis=1))
```
